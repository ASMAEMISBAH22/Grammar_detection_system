```python
# Download and load the "multilingual-gec" dataset from Hugging Face
from datasets import load_dataset
import pandas as pd
dataset = load_dataset("juancavallotti/multilingual-gec")
# Show the first 20 examples from the training split
df = pd.DataFrame(dataset["train"])
df.head(20)
```

```python
# Keep only English (en) and French (fr) sentences
df = df[df["lang"].isin(["en", "fr"])]

# Show how many rows are left after filtering
print("Number of sentences after filtering:", len(df))

# Display the first 20 rows to confirm
df.head(20)
```

```python
# Rename the columns so they make more sense for training
df = df.rename(columns={
    "modified": "source",   # sentence with error
    "sentence": "target"    # correct sentence
})

# Keep only useful columns
df = df[["source", "target", "lang"]]

# Show the first rows to confirm
df.head()
```

```python
# Remove rows that have empty values (NaN)
df = df.dropna()

# Remove duplicate rows
df = df.drop_duplicates()

print("Rows after cleaning:", len(df))
```

```python
import re, html, unicodedata
from ftfy import fix_text

# ------------------------------
# 1. Fix broken Unicode characters
# ------------------------------
# Sometimes sentences in the dataset have weird or broken characters due to encoding issues.
# `fix_text` from ftfy library automatically detects and fixes common Unicode problems,
# making characters display correctly.
df["source"] = df["source"].map(fix_text)
df["target"] = df["target"].map(fix_text)

# ------------------------------
# 2. Decode HTML entities
# ------------------------------
# Some sentences may have HTML codes like &amp; (for &) or &lt; (for <).
# `html.unescape` converts these HTML codes back to normal characters.
# Example: "Hello &amp; goodbye" → "Hello & goodbye"
df["source"] = df["source"].map(html.unescape)
df["target"] = df["target"].map(html.unescape)

# ------------------------------
# 3. Normalize Unicode characters to a standard form
# ------------------------------
# Unicode has multiple ways to represent the same character.
# `unicodedata.normalize("NFKC", x)` converts all characters to a consistent format,
# so the model sees "é" the same way every time.
df["source"] = df["source"].map(lambda x: unicodedata.normalize("NFKC", x))
df["target"] = df["target"].map(lambda x: unicodedata.normalize("NFKC", x))

# ------------------------------
# 4. Collapse multiple spaces and remove extra spaces at the ends
# ------------------------------
# Sometimes sentences have double or triple spaces, or spaces at the start/end.
# We replace any sequence of whitespace with a single space and strip the ends.
# Example: "Hello   world  " → "Hello world"
df["source"] = df["source"].map(lambda x: re.sub(r"\s+", " ", x).strip())
df["target"] = df["target"].map(lambda x: re.sub(r"\s+", " ", x).strip())

# ------------------------------
# 5. Remove spaces before punctuation marks
# ------------------------------
# Some sentences may have a space before punctuation, which is unnatural.
# We remove spaces before . , ! ? ; :
# Example: "Hello world !" → "Hello world!"
df["source"] = df["source"].map(lambda x: re.sub(r"\s+([.,!?;:])", r"\1", x))
df["target"] = df["target"].map(lambda x: re.sub(r"\s+([.,!?;:])", r"\1", x))
```

```python
#Remove rows where source == target (no correction needed)
df = df[df["source"] != df["target"]]
df = df.reset_index(drop=True)
```

```python
#Split dataset into train/valid/test
#80% train, 10% validation, 10% test
#Stratify by language to keep English/French balance

from sklearn.model_selection import train_test_split
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["lang"])
valid_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df["lang"])
```

```python
import os

# Make sure the folder exists
os.makedirs(project_path + "data/processed", exist_ok=True)

# Filter by language and save CSVs
# Train data
train_en_df = train_df[train_df['lang'] == 'en'][["source", "target"]]
train_fr_df = train_df[train_df['lang'] == 'fr'][["source", "target"]]
train_en_df.to_csv(project_path + "data/processed/train_en.csv", index=False, encoding="utf-8")
train_fr_df.to_csv(project_path + "data/processed/train_fr.csv", index=False, encoding="utf-8")

# Validation data
valid_en_df = valid_df[valid_df['lang'] == 'en'][["source", "target"]]
valid_fr_df = valid_df[valid_df['lang'] == 'fr'][["source", "target"]]
valid_en_df.to_csv(project_path + "data/processed/valid_en.csv", index=False, encoding="utf-8")
valid_fr_df.to_csv(project_path + "data/processed/valid_fr.csv", index=False, encoding="utf-8")

# Test data
test_en_df = test_df[test_df['lang'] == 'en'][["source", "target"]]
test_fr_df = test_df[test_df['lang'] == 'fr'][["source", "target"]]
test_en_df.to_csv(project_path + "data/processed/test_en.csv", index=False, encoding="utf-8")
test_fr_df.to_csv(project_path + "data/processed/test_fr.csv", index=False, encoding="utf-8")

print("CSV files saved successfully!")
```
